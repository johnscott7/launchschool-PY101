Question 2: How can you determine whether a given string ends with an exclamation mark (!)? Write some code that prints True or False depending on whether the string ends with an exclamation mark.

```python
str1 = "Come over here!"  # True
str2 = "What's up, Doc?"  # False

# My solution
if str1[-1] == '!':
    print('True')
else:
    print('False')
if str2[-1] == '!':
    print('True')
else:
    print('False')


# Launchschool solution:
print(str1.endswith("!"))
print(str2.endswith("!"))
```

Question 5: Starting with the string `munsters_description = "The Munsters are creepy and spooky."`, print the string with the case of all letters swapped. That is, lowercase letters are converted to uppercase, and uppercase letters are converted to lowercase.
