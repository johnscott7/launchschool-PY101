Question 1: Write two different ways to remove all of the elements from the following list:

```python
numbers = [1, 2, 3, 4]

for i in range(0, len(numbers)):
    numbers.pop()

print(numbers)

numbers2 = [1, 2, 3, 4]
numbers2.clear()
print(numbers2)

numbers3 = [1, 2, 3, 4]
del numbers3[:]
print(numbers3)
```

Question 4: What will the following code output?

```python
my_list1 = [{"first": "value1"}, {"second": "value2"}, 3, 4, 5]
my_list2 = my_list1.copy()
my_list2[0]['first'] = 42
print(my_list1)

# This is a shallow copy which means only the outer list is copied, and the inner lists are shared references. 
# Therefore, my_list2[0] changes the shared reference for my_list as well
# Will print:
# [{"first": 42}, {"second": "value2"}, 3, 4, 5]
```
